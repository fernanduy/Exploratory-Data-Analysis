```python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#edit parameter default matplotlib
from matplotlib import rcParams
rcParams ["figure.figsize"] = 15,8
rcParams ["lines.linewidth"] = 3
rcParams ["xtick.labelsize"] = "x-large"
rcParams ["ytick.labelsize"] = "x-large"
```

```python
#baca dataset
df = pd.read_csv("botak_kotor.csv")
df.shape
```

```python
df = pd.read_csv("botak_kotor.csv").sample(1000)
df.sample(10)
```

```python
#liat kondisi data
df.info()
```

```python
#drop missing data
df.isna().sum()
```

```python
#bagi dataset terbagi menjadi tipe data numerical dan categorical
cat = ["jenis_kelamin", "pekerjaan", "provinsi", "sampo", "pendidikan"]
num = ["umur", "gaji", "is_menikah", "is_keturunan", "berat", "tinggi", "is_merokok", "stress", "botak_prob"]
```

```python
#tampilkan statistika deskriptif
df[num].describe()
```

```python
df[cat].describe()
```

```python
#value counting
for col in cat :
  print (f'''value countkolom (col):''')
  print (df[col].value_counts())
  print()
```

```python
#tampilkan boxplot
for i in range(0, len(num)):
  plt.subplot(1, len(num), i+1)
  sns.boxplot(df[num[i]], color="red", orient="v")
  plt.tight_layout()
```

```python
#tampilkan distribusi plot / violin plot
for i in range(0, len(num)):
  plt.subplot(1, len(num), i+1)
  sns.violinplot(df[num[i]], color="yellow", orient="v")
  plt.tight_layout()
```

```python
#tampilkan barplot / countplot
for i in range(0, len(cat)):
  plt.subplot(2, 3, i+1)
  sns.countplot(df[cat[i]], color="green", orient="v")
  plt.tight_layout()
```

```python
#tampilkan korelasi datanya
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), cmap="Blues", annot=True, fmt=".2")
```

```python
#tampilkan pairplot dimana terdapat target dataset
plt.figure(figsize=(10,10))
sns.pairplot(df, diag_kind='kde')
```

```python
#tampilkan category plot
fig = plt.figure(figsize=(10,10))
for i in range (0, len(num)):
  ax = fig.add_subplot(3,3,i+1)
  sns.catplot(ax=ax, data=df, x="jenis_kelamin", y=num[i])
  plt.tight_layout()
```

```python
fig = plt.figure(figsize=(10,10))
for i in range (0, len(num)):
  ax = fig.add_subplot(3,3,i+1)
  sns.catplot(ax=ax, data=df, x="pekerjaan", y=num[i])
  plt.tight_layout()
```

```python
fig = plt.figure(figsize=(10,10))
for i in range (0, len(num)):
  ax = fig.add_subplot(3,3,i+1)
  sns.catplot(ax=ax, data=df, x="provinsi", y=num[i])
  plt.tight_layout()
```

```python
fig = plt.figure(figsize=(10,10))
for i in range (0, len(num)):
  ax = fig.add_subplot(3,3,i+1)
  sns.catplot(ax=ax, data=df, x="sampo", y=num[i])
  plt.tight_layout()
```

```python
fig = plt.figure(figsize=(10,10))
for i in range (0, len(num)):
  ax = fig.add_subplot(3,3,i+1)
  sns.catplot(ax=ax, data=df, x="pendidikan", y=num[i])
  plt.tight_layout()
```
